```julia
#=Probemos el método de Newton, este se puede llamar dando como argumentos, la función, su derivada y la condición inicial =#

newtonsMethod(x->(x^2-1),x->2*x,60)
```

```julia
#=O también solo dando como argumento la función y la condición inicial=#
newtonsMethod(x->(x^2-1),60)
```

```julia
#=Probemos ahora el método de Newton para un intervalo=#

interval = linspace(-5.,5.,50)

newtonsMethodInterval(x->(x-3)*(x+3),interval)
```

```julia
#=Y si no queremos la larga lista simplemente pedimos que nos escupa las raices. =#
newtonsMethodSpitRoots(x->(x-3)*(x-2)*(x+3),interval,0.01)
```

```julia
#=Por último checamos nuestras funciones para resolver ecuaciones diferenciales, podemos resolver algo simple como dx/dt = -5x 
que sabemos es una exponencial. Si ponemos como condicion inicial x(0) = 10, entonces la exponencial empieza en 10 y decae desde 10=#
    using PyPlot
    plt = PyPlot

    eulerSolucion = eulersMethod((x,t)->-5*x,0,10,10,0.1)
    eulerImplicitoSolucion = implicitEuler((x,t)->-5*x,0.,10.,10.,0.1)
    rungeSolucion = rungeKutta4((x,t)->-5*x,0,10,10,0.1)
 
    plt.xlabel("x")
    plt.ylabel("t")
    plt.title("Solucion a dx/dt = -5x por distintos metodos h =0.1")
    plt.grid()
    plt.tight_layout()
    plt.plot(eulerSolucion[1],eulerSolucion[2],label ="solucion con Euler",linewidth =2) 
    plt.plot(eulerImplicitoSolucion[1],eulerImplicitoSolucion[2],label = "Solucion con Euler implicito",linewidth = 2)
    plt.plot(rungeSolucion[1],rungeSolucion[2],label = "solucion con Runge Kutta")


    legend(bbox_to_anchor=[1.05,1],loc=2,borderaxespad=0)
```
